# Question 22

```python
import numpy as np

def normalize(lis):
    array = np.array(lis)
    maxi = array.max()
    mini = array.min()

    if maxi == mini:
        return [0.5] * len(lis)
    normalized = (array - mini) / (maxi - mini)
    normalized = normalized.tolist()
    normalized.sort()
    return normalized

The_list = [int(i) for i in input().strip().split(",")]

print(f"The values of the list normalized are {normalize(The_list)}")
Normalized_list = normalize(The_list)

import matplotlib.pyplot as plt

plt.plot(Normalized_list)
plt.show()
```

```python
def grade_calculator(num):
    if 100 > num >= 90 :
        print("A")
    if 90 > num >= 80 :
        print("A-")
    if 80 > num >= 70 :
        print("B")
    if 70 > num >= 60 :
        print("B-")
    if 60 > num >= 50 :
        print("C")
    if 50 > num:
        print("Fail")

marks = int(input("Please enter the marks of the student:"))
if 100 >= marks :
    grade_calculator(marks)
else:
    print("Kindly check the marks.")
```
